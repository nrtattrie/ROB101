# Plotting in Julia

Visualizing results is very important as engineers. Code and mathematical output are very hard to explain to people. Graphs and plots are great tools for both showing the significance of your work and debugging!

Julia implements its plots using the ```Plots.jl``` library, which in turn is a metapackage (wrapper) for other plotting Backends. So we can use different plotting libraries with the same syntax. 
The default backend used for plotting in Julia is the GR Backend. 

## Line Plots
The most basic plots in Julia are line plots, lets see a simple exmaple of this. 


````julia
using Plots

x = -pi:0.1:pi;
y = sin.(x)

plot(x,y)
````

### Multiple Plots (Series)

What if we wanted to plots two series of data together? 

Julia interpretes different columns as different series in plotting. So the number of columns in y are different series.
Lets see an example of this

````julia
x = -pi:0.1:pi;
# vector concatenation may be a little counterintuitive, lets also see the size of y to connfirm it has two columns
y = [sin.(x) cos.(x)]
display(size(y))
plot(x,y)
````

### Using Plot!()

Another way of adding another series to your plot is using the plot!(), which returns a handle to current plot object, which you can add the next series to just like you would create a new plot.

````julia
z = cos.(x.+pi/2)
plot!(x,z)
````

## Plot Attributes

Now that we know how to make plots, its important to make sure we know how to set different plot attributes to get better looking and more meaningful plots. 

For instance, in the previous example, we plotted three series together, but the legend (the box that says what color is what) named the series as y1, y2, y3. A more meaningful label would be cos(x) for blue, sin(x) for red, and so on. 

Lets see how to set different attributes for our plots and observe the difference!


````julia
plot(x,y,label = ["sine" "cosine"], linewidth = 3)
````

### Name x, y axis and plot

````julia
plot!(xlabel = "X (rad)", ylabel = "Y", title = "Line Plot Example")
````

````julia
using LaTeXStrings
xtick_range = [-pi,-pi/2,0,pi/2,pi];
xtick_names = [L"-\pi",L"-\pi/2","0",L"\pi/2",L"\pi"]
xticks!((xtick_range,xtick_names))
````

## Scatter Plots

Another most commonly used type of plot is the scatter plot. In the examples above the default plot used by Julia was the lineplot, which joins the different points the series and gives a line plot. However, if we want to see different points in series, we use scatter plots.

You should remember we used a lot of scatter plots in Project 1 for plotting our point clouds! 


````julia
using Plots
default(
    markerstrokewidth=0.3,
    markerstrokecolor=:auto,
    alpha=0.8,
    label=""
)
````

#### Above is another way of specifying plot attributes that stay default in all subsequent plots unless specified otherwise!

````julia
plot(x,y, seriestype = :scatter)
````

````julia
# scatter is also a function in itslef

scatter!(x,z,xlabel = "X (rad)", ylabel = "Y", title = "Scatter Plot Example")
````

## Subplots

A lot of the times, we wish to create a plot side by side instead of plotting them on the same plot. This is usually when the scale of plot is different or the x axis is not the same. We can do this using subplots


````julia
y3 = tan.(x)
p1 = plot(x,[y y3])
p2 = plot(x,y,label = ["sine" "cosine"], linewidth = 3,xlabel = "X (rad)", ylabel = "Y", title = "Fixed plot")
p3 = plot(x,y3, label = "tan", linewidth = 3,xlabel = "X (rad)", ylabel = "Y", title = "Fixed plot")
x_rand = 1:10
y_rand = rand(10,3)
p4 = scatter(x_rand,y_rand, title = "random scatter plot")
plot(p1, p2, p3, p4, layout = (2, 2))
````
